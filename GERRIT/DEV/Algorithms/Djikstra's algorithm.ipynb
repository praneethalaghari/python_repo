```python
import math

class Node:
    
    
    def __init__(self,name):
        self.__name = name
        self.adjacencylist = {}
        
    def getName(self):
        return self.__name
    
    def setXdistance(self,X):
        self.xdistance = X
    
    def setYdistance(self,Y):
        self.ydistance = Y
        
    def getXdistance(self):
        return self.xdistance
    
    def getYdistance(self):
        return self.ydistance
    
    def getcoordinates(self):
        return self.getXdistance(),self.getYdistance()
    
    #Here the values are weights of vertices i.e distances 
    def add_adjacent_node(self,node):
        self.adjacencylist[node.getName()] = math.sqrt(pow(self.getXdistance() - node.getXdistance(),2) +  pow(self.getYdistance() - node.getYdistance(),2))
        
class Graph:
    network = []
    
    def add_node(self,node):
        Graph.network.append(node)
  
    def get_network(self):
        return Graph.network
    
    def print_nodes(self):
        print([node.getName() for node in Graph.network])
    
    
    def print_graph(self):
        for node in Graph.network:
            print(node.getName(),"->".join([x for x in node.adjacencylist.keys()]),sep='->')
            
    def print_vertices(self):
        for node in Graph.get_network(Graph):
            print(node.getName(),dict(node.adjacencylist),sep = "\t")
        
        
if __name__ == '__main__':
    
    g = Graph()
    
    #Nodes
    
    n1 = Node('A');n1.setXdistance(1);n1.setYdistance(2)
    g.add_node(n1)
    
    n2 = Node('B');n2.setXdistance(5);n2.setYdistance(3)
    
    g.add_node(n2)
    
    n3 = Node('C');n3.setXdistance(9);n3.setYdistance(3)
    
    g.add_node(n3)
    
    n4 = Node('D');n4.setXdistance(14);n4.setYdistance(3)
    
    g.add_node(n4)
    
    n5 = Node('E');n5.setXdistance(3);n5.setYdistance(1)
    
    g.add_node(n5)
    
    n6 = Node('F');n6.setXdistance(8);n6.setYdistance(1)
    
    g.add_node(n6)
    
    n7 = Node('G');n7.setXdistance(13);n7.setYdistance(0)
    
    g.add_node(n7)
    
    n8 = Node('H');n8.setXdistance(17);n8.setYdistance(1)
    
    g.add_node(n8)
    
    n9 = Node('I');n9.setXdistance(19);n9.setYdistance(2)
    
    g.add_node(n9)
    
    
    #Vertices
    
    n1.add_adjacent_node(n2);n2.add_adjacent_node(n1)     # Undirected Graph i.e A adjacent to B is B adjacent to A if directed then no need
    n1.add_adjacent_node(n5);n5.add_adjacent_node(n1)
    n2.add_adjacent_node(n5);n5.add_adjacent_node(n2)
    n2.add_adjacent_node(n6);n6.add_adjacent_node(n2)
    n5.add_adjacent_node(n3);n3.add_adjacent_node(n5)
    n4.add_adjacent_node(n3);n3.add_adjacent_node(n4)
    n4.add_adjacent_node(n6);n6.add_adjacent_node(n4)
    n4.add_adjacent_node(n9);n9.add_adjacent_node(n4)
    n8.add_adjacent_node(n9);n9.add_adjacent_node(n8)
    n7.add_adjacent_node(n3);n3.add_adjacent_node(n7)
    n7.add_adjacent_node(n6);n6.add_adjacent_node(n7)
    n7.add_adjacent_node(n8);n8.add_adjacent_node(n7)
    
    
    g.print_nodes()
    g.print_graph()
    g.print_vertices()
    
    
    
```
