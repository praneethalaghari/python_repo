# MATPLOTLIB

## Basic Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4])
b = np.array([3,4,5,6])
c = np.array([5,6,7,8])
d = np.linspace(0,9,19)
e = np.logspace(1,5,19)
d,e
```

```python
plot = plt.plot(a,a*2,'blue')     # Self plotting i.e use only one element of 'a' array for x and y axis
```

```python
plot2 = plt.plot(a,b-a*2,'red')    # Plotting using multiple arrays i.e a for x axis,b for y axis 
```

```python
plot3 = plt.plot(a,b,c,'green')  # If we give 3 arrays then two lines will be plotted. i.e 'a' against 'b';'a' against 'c'
```

## Sub Plots

```python
# This divides the canvas area into 2X2 matrix and each part is accessed by 3rd element i.e index;
plt.subplot(2,2,1)     # (2,2,1) is top left
plt.plot(a,b,'g')

sbplt = plt.subplot(2,2,4)     # (2,2,4) is bottom right
sbplt.set_label('New')
plt.plot(a,b,'r')
```
