### Figure plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.randint(7,size=(8,))
b = np.random.randn(a.size).reshape(a.shape)
```

```python
c = np.array([[1,2,3]])
d = np.array([4,5,6])
c.shape,c[0].shape,c.ndim,c[0].ndim
```

```python
fig = plt.figure(figsize=(16,9))
axes = fig.add_axes([0,0,1,1])    # Generally used in case we dont intend to use sub plots
axes.plot(a,b,'r')
```

```python
e = np.array([[1,2,3],[4,5,6]])
e.ndim,e.shape
```

```python
plt.plot(e[0],e[1],'y')     # OMG!! - Another type of self plotting  :) i.e via 2 dimensional array
```

```python
fig.suptitle("Fig plot",fontsize=20)   # Naming the plot
fig
```
