## Figure Subplots and decoration

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This is an alternative way of creating multiple plots.. i.e we can create multiple plots via add_axes
'''
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0,0,1,1])
axes1.plot()
axes2.plot()
'''

fig,axes = plt.subplots(1,2,figsize=(8,4.5))    # Here axes will contains all the subplot elements
axes0, = axes[0].plot([1,2,3],[1,2,3],color='red',alpha=0.5,lw='3',marker='o')
axes1, = axes[1].plot([1,2,3],[3,2,1],color='maroon',alpha=0.5,lw='3',marker='*',markersize='10')
type(axes[0]),type(axes0)
```

```python
fig.suptitle("Figure",fontsize='20')

axes[0].set_title("Fig 1",fontsize=15)
axes[1].set_title("Fig 2",fontsize=15)

axes[0].set_xlabel("X-AXIS",fontsize='15')
axes[0].set_ylabel("Y-AXIS",fontsize='15')

axes[1].set_xlabel("X-AXIS",fontsize='15')
axes[1].set_ylabel("Y-AXIS",fontsize='15')

#type(axes[0]) = matplotlib.axes._subplots.AxesSubplot
#type(axes0) =  matplotlib.lines.Line2D

#Axes dont have color,alpha,marker etc.. attributes
#Line2D have attributes hence access them.. here axes1 (not axes[1]) is Line 2D object not axes) object

axes1.set_color('green')
axes1.set_alpha(0.5)

fig.tight_layout(rect=[0,0,1,1])                           # This adjusts the subplots alignment considering Labels,Texts
fig
```
