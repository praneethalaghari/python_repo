# BAR, SCATTER, FILL_BETWEEN, PIE, STEP, HISTOGRAMS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4])
b = np.array([1,2,3,4])
```

```python
fig,axes = plt.subplots()
axes.bar(a,b)
```

```python
fig,axes = plt.subplots()
axes.scatter(a,b)
```

```python
fig,axes = plt.subplots()
axes.fill_between(a,b)
axes.grid(True)      # To have grids
```

```python
fig,axes = plt.subplots()
axes.pie(a)
```

```python
fig,axes = plt.subplots()
axes.step(a,b)
```

```python
fig,axes = plt.subplots()
axes.hist(a,b)
```
